```python
import pandas as pd 

df = pd.read_csv(r'D:\FPTUniversity\Semester 8\DSP391m\VN-Index\Raw Data\VN Index Historical Data(daily).csv')
df
```

```python

df['Price'] = df['Price'].str.replace(',', '').astype(float)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'] )
df_sorted = df.sort_values(by= 'Date', ascending= True)
df = df_sorted
df = df.reset_index(drop= True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse
import matplotlib.pyplot as plt 


# Multiplicative Decomposition 
multiplicative_decomposition = seasonal_decompose(df['Price'], model='multiplicative', period=30)

# Additive Decomposition
additive_decomposition = seasonal_decompose(df['Price'], model='additive', period=30)

# Plot
plt.rcParams.update({'figure.figsize': (16,12)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

additive_decomposition.plot().suptitle('Additive Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```
