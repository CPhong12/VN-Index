### Import library

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv(r"C:\Code\VN-Index\Raw Data\Giá vàng,CPI, tỉ giá hối đoái.csv", encoding='ISO-8859-1')
df_1
```

### Xử lí bảng df1

```python
# Reshape dataframe từ dạng wide thành dạng long
df_melted = df_1.melt(id_vars=['Nam', 'Nhom hang'], var_name='Thang', value_name='Gia tri')

# Chuyển đổi cột 'Thang' thành số nguyên
df_melted['Thang'] = df_melted['Thang'].astype(int)

# Tạo cột 'Thang/Nam'
df_melted['Thoi gian'] = 'Tháng ' + df_melted['Thang'].astype(str) + '/' + df_melted['Nam'].astype(str)

df_melted
```

```python
# Chuyển dấu phẩy thành dấu chấm và chuyển thành float
df_melted['Gia tri'] = df_melted['Gia tri'].astype(float)

# Dùng pivot để chuyển từ dạng long sang wide
df_wide = df_melted.pivot(index='Thoi gian', columns='Nhom hang', values='Gia tri')

df_wide.reset_index(inplace=True)

df_wide
```

```python
df_wide['Thoi gian'] = df_wide['Thoi gian'].str.replace('Tháng ', '', regex=False)

df_wide
```

```python
df_wide.to_csv("C:\Code\VN-Index\Project\Data preprocessing\Data\Data_preprocessing\Price.csv", index=False)
```
