### Import Lib & DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- df = bảng các chỉ số vĩ mô
- df1 = bảng của chứng khoáng

```python
df = pd.read_csv('C:\Code\VN-Index\Project\Data preprocessing\Data\Giá.csv')

df.head()
```

```python
df1 = pd.read_csv('C:\Code\VN-Index\Raw Data\VN Index Historical Data (months).csv')

df1.head()
```

## Visualize giữa các chỉ số Vĩ Mô

### Biểu đồ scatter plot cho từng cặp chỉ số

```python
plt.figure(figsize=(10, 5))
sns.pairplot(df[['Chi so do la My', 'Chi so gia tieu dung', 'Chi so gia vang']],plot_kws={'color': 'green'})
plt.suptitle('Scatter Plot giữa các chỉ số',y=1.02)
plt.show()
```

### Vẽ heatmap tương quan


```python
plt.figure(figsize=(8, 6))
correlation_matrix = df[['Chi so do la My', 'Chi so gia tieu dung', 'Chi so gia vang']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap thể hiện hệ số tương quan giữa các chỉ số')
plt.show()
```

## Visualize giữa các chỉ số Vĩ mô so với chứng khoáng

```python
plt.figure(figsize=(12, 6))

sns.lineplot(data=df, x='Thoi gian', y='Chi so do la My', label='Chỉ số đô la Mỹ', marker='o')
sns.lineplot(data=df, x='Thoi gian', y='Chi so gia tieu dung', label='Chỉ số giá tiêu dùng', marker='o')
sns.lineplot(data=df, x='Thoi gian', y='Chi so gia vang', label='Chỉ số giá vàng', marker='o')
sns.lineplot(data=df1, x='Date', y='Price', label='Giá cổ phiếu', marker='o', linestyle='--')

plt.title('Biểu đồ biến động giữa các chỉ số vĩ mô và giá cổ phiếu theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Giá trị')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
