```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("C:\Code\VN-Index\Project\Data preprocessing\Data\Data_preprocessing\Giá dầu.csv")
df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
df['Date'] = df['Date'].str.replace('Tháng ', '')
df['Date']=pd.to_datetime(df['Date'])

df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Value'], label='Million USD')
plt.title('Oil Price Chart Over Time')
plt.xlabel('Months')
plt.ylabel('Price')
plt.xticks()
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
