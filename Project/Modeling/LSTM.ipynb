```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
vni_daily = pd.read_csv(r'D:\FPTUniversity\Semester 8\DSP391m\VN-Index\Raw Data\VN Index Historical Data(daily).csv')
vni_daily
```

```python
# Đảm bảo cột 'Date' ở dạng datetime
vni_daily['Date'] = pd.to_datetime(vni_daily['Date'], format='%m/%d/%Y')

# Sắp xếp theo cột 'Date' theo thứ tự tăng dần
vni_sorted = vni_daily.sort_values(by='Date', ascending=True)
vni_daily = vni_sorted
vni_daily= vni_daily.reset_index(drop= True)
vni_daily 
```

```python
vni_daily['Price'] = vni_daily['Price'].str.replace(',', '').astype(float)
vni_daily['Open'] = vni_daily['Open'].str.replace(',', '').astype(float)
vni_daily['High'] = vni_daily['High'].str.replace(',', '').astype(float)
vni_daily['Low'] = vni_daily['Low'].str.replace(',', '').astype(float)
vni_daily
```

```python
vni_daily['Date'] = pd.to_datetime(vni_daily['Date'])
filtered_df = vni_daily[vni_daily['Date'] < '2023-01-01']
test_df = vni_daily[vni_daily['Date'] > '2022-12-31']
print(test_df)
```

```python
cols_name = list(vni_daily)[2:5]
df_for_training = filtered_df[cols_name]
df_for_testing = test_df[cols_name]
print(df_for_training)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
df_for_training_scaled = scaler.transform(df_for_training)
df_for_testing_scaled = scaler.transform(df_for_testing)
```

```python
train_X = []
train_Y = []

n_future = 1
n_past = 14

for i in range(n_past, len(df_for_training_scaled) - n_future + 1):
    train_X.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
    train_Y.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

train_X, train_Y = np.array(train_X), np.array(train_Y)

print('trainX shape == {}.'.format(train_X.shape))
print('trainY shape == {}.'.format(train_Y.shape))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(train_X.shape[1], train_X.shape[2]), return_sequences=True))
model.add(LSTM(32, activation='relu', return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(train_Y.shape[1]))

model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history = model.fit(train_X, train_Y, epochs=10, batch_size=16, validation_split=0.15, verbose=1)

plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
```

```python
test_X = []
test_Y = []

n_future = 1
n_past = 14

for i in range(n_past, len(df_for_testing_scaled) - n_future + 1):
    test_X.append(df_for_testing_scaled[i - n_past:i, 0:df_for_testing.shape[1]])
    test_Y.append(df_for_testing_scaled[i + n_future - 1:i + n_future, 0])

test_X, test_Y = np.array(test_X), np.array(test_Y)

print('testX shape == {}.'.format(test_X.shape))
print('testY shape == {}.'.format(test_Y.shape))
```

```python

n_past = 14  
n_features = 3  
n_steps = 5  


def sliding_window_prediction(model, data, window_size=n_past, n_steps=n_steps):
    predictions = []
    window_data = data[-window_size:]  

    
    for _ in range(n_steps):
       
        pred = model.predict(window_data.reshape(1, window_size, n_features))

        
        predictions.append(pred[0][0])

        
        high_variation = np.random.uniform(0, 0.07)  
        low_variation = np.random.uniform(0, 0.07) 
        high = pred[0][0] * (1 + high_variation)  
        low = pred[0][0] * (1 - low_variation)   


        
        open_price = (high + low) / 2

        
        new_window_data = np.array([open_price, high, low]).reshape(1, n_features)
        
        
        window_data = np.roll(window_data, shift=-1, axis=0)  
        window_data[-1] = new_window_data  

    return np.array(predictions)
```

```python
prediction = model.predict(test_X)
#prediction = model.predict(train_X)
prediction_copies = np.repeat(prediction, df_for_testing.shape[1], axis=-1)
y_pred_future = scaler.inverse_transform(prediction_copies)[:, 0]
```

```python
test_Y_copies = np.repeat(test_Y, df_for_testing.shape[1], axis=-1)
test_Y_inverse = scaler.inverse_transform(test_Y_copies)[:, 0]
```

```python
test_Y_draw = test_Y_inverse.flatten()
prediction_draw = y_pred_future.flatten()


plt.figure(figsize=(12,8))
# Tạo các chỉ số (x-axis)
x = np.arange(len(test_Y_draw))

# Vẽ biểu đồ
plt.plot(x, test_Y_draw, label='Thực tế', color='b')
plt.plot(x, prediction_draw, label='Dự đoán', color='r')

# Thêm nhãn và tiêu đề
plt.xlabel('Index')
plt.ylabel('Giá trị')
plt.title('So sánh Thực tế và Dự đoán')

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

```python
rmse = np.sqrt(np.mean(((prediction - test_Y) ** 2)))
mae = np.mean(np.abs(prediction - test_Y))
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```

```python


future_predictions = sliding_window_prediction(model, df_for_testing_scaled, n_steps=n_steps)


future_predictions_expanded = np.hstack([future_predictions.reshape(-1, 1), np.zeros((n_steps, n_features - 1))])
future_predictions_original = scaler.inverse_transform(future_predictions_expanded)[:, 0]  

print("Dự đoán cho 5 ngày tiếp theo:", future_predictions_original)
    
```
