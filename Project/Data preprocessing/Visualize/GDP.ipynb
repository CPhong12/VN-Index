```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("C:\Code\VN-Index\Project\Data preprocessing\Data\Data_preprocessing\GDP.csv")
```

```python
filtered_df = df[~df['Date'].str.contains("tháng")]
```

```python
filtered_df.head()
```

```python
def convert_quarter_to_date(quarter_str):
    quarter, year = quarter_str.split('/')
    if '1' in quarter:
        return pd.Timestamp(f'01-01-{year}')
    elif '2' in quarter:
        return pd.Timestamp(f'01-04-{year}')
    elif '3' in quarter:
        return pd.Timestamp(f'01-07-{year}')
    elif '4' in quarter:
        return pd.Timestamp(f'01-10-{year}')

# Áp dụng hàm chuyển đổi cho cột 'Date'
filtered_df['Date'] = filtered_df['Date'].apply(convert_quarter_to_date)

# Vẽ biểu đồ cột
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Date', y='Value', palette='viridis')
plt.title('Biểu đồ cột GDP theo Quý')
plt.xlabel('Thời gian (Quý)')
plt.ylabel('GDP (Tỷ VNĐ)')
plt.xticks(rotation=45)
plt.grid(axis='y')  # Hiển thị lưới theo trục y
plt.tight_layout()
plt.show()
```
